```python
import numpy as np
import IPython.display as ipd
import librosa as lb
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

```python
def generate_signal(feq,time,am,phase = 0):
    signal = am*np.sin(2*np.pi*feq*time + phase)
    return signal
    
```

```python
generate_signal(100,np.linspace(0,10,200),50,0)
```

```python
np.pi
```

```python
def plot_signal(signal,time):
    plt.figure(figsize=(10,7))
    plt.xlabel("Time")
    plt.ylabel("Signal")
    plt.plot(time,signal)
    plt.show()
```

```python
sampling_rate = 44100
T = 50
n_rate = int(sampling_rate/2)
time = np.linspace(0,T,T*n_rate)
signal = generate_signal(10,time,50,40)
plot_signal(signal,time)
```

# ADDING SIGNALS TO GENERATE A NEW SIGNAL

```python
sig_1 = generate_signal(100,time,50,40)
sig_2 =  generate_signal(200,time,30,0)
sig_3 =  generate_signal(300,time,20,30)
sig_4 =  generate_signal(400,time,20,33)
sig_5 =  generate_signal(500,time,10,0)
total_sig = sig_1 + sig_2 + sig_3 + sig_4 + sig_5
```

```python
plot_signal(total_sig,time)
```

```python

ipd.Audio(total_sig,rate = sampling_rate)
```
